````python
# Nested Loop Example 1: No functions

colorList = ["red", "blue", "green", "yellow", "purple", "orange", "maroon", "aqua", "teal"]
index_1 = 0
index_2 = 0

e_count = 0
vowel_count = 0


for index_1 in range(len(colorList)):
  current_color = colorList[index_1]
  for index_2 in range(len(current_color)):
    print("index 1: " + str(index_1) + ", index 2: " + str(index_2))
    letter = current_color[index_2]
    print("The current color is " + current_color + " and the current letter is " + letter)
    if letter == "e":
      e_count += 1
      vowel_count +=1
      print("We've found an e! There are " + str(e_count)+ " e's so far, and "+ str(vowel_count)+" total vowels in colorList")
    elif (letter == "a" or letter == "i" or letter == "o" or letter == "u"):
      vowel_count += 1
      print("we've found a vowel! There are " + str(vowel_count) + " vowels so far")
    else:
      print(letter + " is not a vowel")

print("\nThe total number of e's in colorList is "+ str(e_count) + " and the total number of vowels in colorList is " + str(vowel_count))
````

````python
# Write a program that prints the first 20 Fibonacci numbers and their sum.

def addFibonacci(fibonacci):
  sum = 0
  for num in fibonacci:
    sum += num
  print(sum)

def getFibonacci(fibonacci):
  for index in range(2, 20):
    nextFibonacci = fibonacci[index - 1] + fibonacci[index - 2]
    fibonacci.append(nextFibonacci)
    print(fibonacci)

if __name__ == "__main__":
  fibonacci = [0, 1]
  print(fibonacci)
  getFibonacci(fibonacci)
  addFibonacci(fibonacci)
````
